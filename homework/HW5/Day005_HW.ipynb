```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
import requests
import json

res = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
results = json.loads(res.text)
print('Ans: æ­¤APIå›å‚³' + str(len(results)) + 'ç­†è³‡æ–™')

for result in results:
    print('====================================')
    print('æ¯ç­†ç¸½å…±æœ‰å¤šå°‘æ¬„ä½(åªç®—ç¬¬ä¸€å±¤):' + str(len(result)))
    for x,y in result.items():
        print(x + "                " + str(y))
# å·®ä¸€å€‹æ¬„ä½ç‚º departmentã€‚

```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
# å°‡æ¯ç­†è³‡æ–™çš„å››å€‹æ¬„ä½ï¼Œä¾åºç”±æ¨™é¡Œã€è²¼æ–‡æ™‚é–“ã€ç•™è¨€äººæ•¸ã€æŒ‰è®šäººæ•¸ï¼Œå­˜ç‚ºä¸€å€‹list object
# å†å°‡è³‡æ–™å­˜æ–¼ans2ä¹‹ä¸­
ans2 = []
for result in results:
    record = []          # å…¨æ–°çš„recordè¨˜éŒ„æ¯ç­†è³‡æ–™å…¶ä¸­çš„å››å€‹æ¬„ä½
    record.append(result['title'])
    record.append(result['createdAt'])
    record.append(result['commentCount'])
    record.append(result['likeCount'])
    ans2.append(record)
    
print(ans2)
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

avgCommentCount = 0.0;
avgLikeCount = 0.0;
totalCommentCount = 0;
totalLikeCount = 0;

for record in ans2:
    totalCommentCount += record[2]
    totalLikeCount += record[3]

avgCommentCount = totalCommentCount / len(ans2)
avgLikeCount = totalLikeCount / len(ans2)
print('ç†±é–€æ–‡ç« çš„å¹³å‡ç•™è¨€äººæ•¸ç‚ºï¼š' + str(avgCommentCount))
print('ç†±é–€æ–‡ç« çš„å¹³å‡æŒ‰è®šäººæ•¸ç‚ºï¼š' + str(avgLikeCount))

# å–å¾—éç†±é–€æ–‡ç« çš„å…§å®¹
unPop = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
unPopDict = json.loads(unPop.text)
for record in unPopDict:
    totalCommentCount += record['commentCount']
    totalLikeCount += record['likeCount']

print('éç†±é–€æ–‡ç« ç¸½æ•¸ï¼š' + str(len(unPopDict)))
avgCommentCount = totalCommentCount / len(unPopDict)
avgLikeCount = totalLikeCount / len(unPopDict)
print('éç†±é–€æ–‡ç« çš„å¹³å‡ç•™è¨€äººæ•¸ç‚ºï¼š' + str(avgCommentCount))
print('éç†±é–€æ–‡ç« çš„å¹³å‡æŒ‰è®šäººæ•¸ç‚ºï¼š' + str(avgLikeCount))
```
