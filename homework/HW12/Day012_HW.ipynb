### ① 取出今天所有的發文 (https://www.ettoday.net/news/news-list.htm)

```python
import requests
from bs4 import BeautifulSoup

target_url = 'https://www.ettoday.net/news/news-list.htm'
resp = requests.get(target_url)
# 以瀏覽器方式解析文檔('html5lib')，還有其它種選擇(e.g. lxml,html.parser,xml)
soup = BeautifulSoup(resp.text,'html5lib')

for news in soup.find('div','part_list_2').find_all('h3'):
    # 取出今天所有發文，以{時間、新聞分類、新聞主題}條列
    print(news.find(attrs={'class':'date'}).text,news.find('em').text, news.find('a').text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
news_dict = {}
for news in soup.find('div','part_list_2').find_all('h3'):
    # 取出今天所有發文，以{時間、新聞分類、新聞主題}
    # 以分類當作Key，相關新聞存為list，再用dict方式儲存
    anews = []
    anews.append(news.find(attrs={'class':'date'}).text)
    anews.append(news.find('a').text)
    if news_dict.__contains__(news.find('em').text):
        temp = news_dict[news.find('em').text]
        temp.append(anews)
        dnews = {news.find('em').text:temp}
        news_dict.update(dnews)
    else:
        dnews = {news.find('em').text:anews}
        news_dict.update(dnews)
    
print('Total News Category Count is {}'.format(len(news_dict.items())))
for item in news_dict.keys():
    print('新聞類別：{}'.format(item))
    print(news_dict[item])
    print('--------------------------')
```

### ③ 哪一個類別的文章最多

```python
max_count_category = ''
max_count = 0
for key in news_dict.keys():
    print(key,len(news_dict[key]))
    if max_count < len(news_dict[key]):
        max_count = len(news_dict[key])
        max_count_category = key
        
print('擁有最多文章類別為[{}]，數量為{}'.format(max_count_category,max_count))
```
