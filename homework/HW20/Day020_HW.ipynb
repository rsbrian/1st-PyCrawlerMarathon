```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
d = dict({})

rows = table.find_all('tr')
keyword = ''
for row in rows:
    temp_dict = dict({})
    cols = row.find_all('td')
    # 依欄讀取資料
    if len(cols) == 3:        # 日期/數字/空白
        temp_dict.update({cols[0].text:cols[1].text})
        if keyword in d:
            tmp = d.get(keyword)
            tmp.update(temp_dict)
        else:
            d.update({keyword:temp_dict})
    elif len(cols) == 5:      # 氣體名稱
        keyword = cols[0].text
    else:
        continue
    
print(d)
```
